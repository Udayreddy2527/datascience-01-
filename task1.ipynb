### Variables Identification

**Independent Variables (Inputs):**
- Genre
- Year
- Duration
- Votes

**Dependent Variable (Output):**
- IMDb Rating

Reason: Movie rating depends on genre, duration, year, and votes.


### Dataset Meaning

This dataset contains information about movies listed on IMDb, including
movie name, genre, release year, duration, number of votes, and IMDb rating.
It is used to analyze factors that influence movie ratings.


### Assumptions
- IMDb ratings are accurate
- Votes are genuine
- Data is correctly recorded

### Limitations
- Dataset is limited to IMDb only
- Missing or biased data may exist
- Audience demographics are not included
